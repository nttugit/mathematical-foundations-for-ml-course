## Plotting a system of linear equations

### Bank robber and sheriff excercise

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 40,100) # start, finish, num of points
```

```python
# distance travel by robber, d = 2.5t
d_r = 2.5 * t
```

```python
# distance travel by sheriff, d = 3(t-5)
d_s = 3 * (t - 5)
```

```python
fig, ax = plt.subplots()

ax.set_title("A Bank Robber Caught")
ax.set_xlabel("Time (minute)")
ax.set_ylabel("Distance (km)")
ax.set_xlim([0,40])
ax.set_ylim([0,100])

ax.plot(t, d_r, c='red', label='Bank Robber')
ax.plot(t, d_s, c='green', label='Sheriff')
plt.axvline(x=30, color='purple', linestyle='--')
plt.axhline(y=75, color='purple',linestyle='--')

plt.show()
```

```python
# If the speed of the bank robber and the sheriff are the same => Never catch
t = np.linspace(0, 40,60) # start, finish, num of points
# distance travel by robber, d = 1.t
d_r2 = 1 * t
# distance travel by sheriff, d = 1.(t-5)
d_s2 = 1 * (t - 5)

fig, ax = plt.subplots()



ax.set_title("A Bank Robber Caught")
ax.set_xlabel("Time (minute)")
ax.set_ylabel("Distance (km)")
ax.set_xlim([0,40])
ax.set_ylim([0,60])

ax.plot(t, d_r2, c='red', label='Bank Robber')
ax.plot(t, d_s2, c='green', label='Sheriff')

plt.show()
```

### Mark I and Mark II design

```python
t = np.linspace(0,60,100)
# kj generated by Mark I design
e_1 = 1 * t
# kj generated by Mark II design
e_2 = 4 * (t - 30)
```

```python
fg, ax = plt.subplots()
ax.set_title('Power generated by Mark I and Mark II designs')
ax.set_xlabel("Time (days)")
ax.set_ylabel("Generated power (kJ)")
ax.set_xlim([0, 60])
ax.set_ylim([0, 100])

plt.axvline(x=40, color='purple', linestyle='--')
plt.axhline(y=40, color='purple', linestyle='--')
ax.plot(t,e_1,c='green', label='Mark I')
ax.plot(t,e_2, c='red', label='Mark II')

plt.legend()
plt.show()
```
